```python
import numpy as np
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
import os
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.manifold import Isomap
def readmat(filepath):
    mat = scipy.io.loadmat(filepath)
    features = mat['fea']
    classes = mat['gnd']
    classes = classes.transpose()[0]
    return (features, classes)

def Redim_on_FCPS(fcps_dir, redim_func):
    files = [os.path.join(fcps_dir, file) for file in os.listdir(fcps_dir)]
    for file in files:
        redim_func(file)
```

# PCA on FCPS

```python
def PCA_on_MAT(Path):
    file_name = Path.split('/')[-1]
    X,classes = readmat(Path)
    pca = PCA(n_components=2)
    X_pca = pca.fit(X).transform(X)

    print(file_name + ' explained variance ratio (first two components): %s'
          % str(pca.explained_variance_ratio_))
    plt.figure()
    plt.scatter(X_pca[:, 0], X_pca[:, 1],
                c=classes, edgecolor='none', alpha=0.5)
    plt.xlabel('component 1')
    plt.ylabel('component 2')
    plt.title(file_name +" PCA")
    plt.colorbar();
```

```python
Redim_on_FCPS("./TP_Etude_de_cas_20/FCPS/", PCA_on_MAT)
```

# MDS on FCPS

```python
def MDS_on_MAT(Path):
    file_name = Path.split('/')[-1]
    X,classes = readmat(Path)
    mds = MDS(n_components=2)
    X_mds = mds.fit_transform(X)

    plt.figure()
    plt.scatter(X_mds[:, 0], X_mds[:, 1],
                c=classes, edgecolor='none', alpha=0.5)
    plt.xlabel('component 1')
    plt.ylabel('component 2')
    plt.title(file_name +" MDS")
    plt.colorbar();
    
```

```python
Redim_on_FCPS("./TP_Etude_de_cas_20/FCPS/", MDS_on_MAT)
```

# LLE on FCPS

```python
def LLE_on_MAT(Path):
    file_name = Path.split('/')[-1]
    X,classes = readmat(Path)
    lle = LocallyLinearEmbedding(n_components=2)
    X_mds = lle.fit_transform(X)

    plt.figure()
    plt.scatter(X_mds[:, 0], X_mds[:, 1],
                c=classes, edgecolor='none', alpha=0.5)
    plt.xlabel('component 1')
    plt.ylabel('component 2')
    plt.title(file_name +" LLE")
    plt.colorbar();
```

```python
Redim_on_FCPS("./TP_Etude_de_cas_20/FCPS/", LLE_on_MAT)
```

# Hessian EIGENMAP on FCPS

```python
def HEIGEN_on_MAT(Path):
    file_name = Path.split('/')[-1]
    X,classes = readmat(Path)
    heigen = LocallyLinearEmbedding(n_components=2, method = 'hessian')
    X_mds = heigen.fit_transform(X)

    plt.figure()
    plt.scatter(X_mds[:, 0], X_mds[:, 1],
                c=classes, edgecolor='none', alpha=0.5)
    plt.xlabel('component 1')
    plt.ylabel('component 2')
    plt.title(file_name +" H_EIGEN")
    plt.colorbar();
```

```python
Redim_on_FCPS("./TP_Etude_de_cas_20/FCPS/", HEIGEN_on_MAT)
```

```python
def Isomap_on_MAT(Path):
    file_name = Path.split('/')[-1]
    X,classes = readmat(Path)
    iso = Isomap(n_components=2)
    X_mds = iso.fit_transform(X)

    plt.figure()
    plt.scatter(X_mds[:, 0], X_mds[:, 1],
                c=classes, edgecolor='none', alpha=0.5)
    plt.xlabel('component 1')
    plt.ylabel('component 2')
    plt.title(file_name +" ISOMAP")
    plt.colorbar();
```

```python
Redim_on_FCPS("./TP_Etude_de_cas_20/FCPS/", Isomap_on_MAT)
```
